```python
#import necessary libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests,json, re
from selenium import webdriver
import numpy as np

#scrape pokemon list website
pokemon_website = "https://pokemondb.net/pokedex/national"
listofpokemon = requests.get(pokemon_website)
soup = BeautifulSoup(listofpokemon.text, 'html.parser')

#create dataframe based on names
pokemonnames = soup.findAll('a', {'class': 'ent-name'})
print(pokemonnames)
df = pd.DataFrame(pokemonnames)
print(df)

#numpy array to allow easier length computations
arrayofnames = df.to_numpy()

lengthofname = int(input("How many letters do you want this pokemon to have in its name?: "))
for name in arrayofnames:
  if len(name[0]) == lengthofname:
    print(name)
```
